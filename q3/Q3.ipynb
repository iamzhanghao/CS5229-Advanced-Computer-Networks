```python
import pandas as pd


data = []

for i in (list(range(0,104,2))+list(range(100,1000,100))):
    for mode in ['bbr','cubic']:
        for flow in ['flow1','flow2']:
            
            f = open(mode + "_" + str(i) + "_" + flow, "r").readlines()
            throughput = float(f[-4].split("ytes")[1].split(' ')[2])

            data.append([mode,i,throughput,flow])
        

        
df = pd.DataFrame(data = data,
                  columns=['mode','delay','throughput','flow'])

# df['total'] = df['throughput_flow1'] + df['throughput_flow2']


        
df.head(300)        
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_bbr = df[df['mode']=='bbr']

sns.set_style("darkgrid")
sns.lineplot(x="delay", dashes=False, y="throughput", hue="flow", style="flow", data=df_bbr)
plt.title("Throughput BBR")
plt.show()

sns.set_style("darkgrid")
sns.lineplot(x="delay", dashes=False, y="throughput", hue="flow", style="flow", data=df_bbr)
plt.title("Throughput BBR - Zoom In")
plt.xlim(0, 100)
# plt.ylim(0, 150)
plt.show()
```

```python
df_cubic = df[df['mode']=='cubic']

sns.set_style("darkgrid")
sns.lineplot(x="delay", dashes=False, y="throughput", hue="flow", style="flow", data=df_cubic)
plt.title("Throughput Cubic")
plt.show()

sns.set_style("darkgrid")
sns.lineplot(x="delay", dashes=False, y="throughput", hue="flow", style="flow", data=df_cubic)
plt.title("Throughput Cubic - Zoom In")
plt.xlim(0, 100)
# plt.ylim(0, 150)
plt.show()
```
