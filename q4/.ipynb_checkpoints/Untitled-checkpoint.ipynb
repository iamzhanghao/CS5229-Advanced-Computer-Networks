```python
import pandas as pd


data = []

for i in (list(range(4,100,4))+list(range(100,3000,50))):
    for flow in ['bbr','cubic']:
        
        f = open("2hop_" + str(i) + "_" + flow, "r").readlines()
        t = f[-4].split("ytes")[1].split()
        if (t[1]=='Mbits/sec'):
            throughput = float(t[0])
        else:
            throughput = float(t[0])/1000.0
        

        data.append([flow,i,throughput])
        

        
df = pd.DataFrame(data = data,
                  columns=['flow','queue_size','throughput'])

# df['total'] = df['throughput_flow1'] + df['throughput_flow2']


df.head(200)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# df_bbr = df[df['flow']=='bbr']

sns.set_style("darkgrid")
sns.lineplot(x="queue_size", dashes=False, y="throughput", hue="flow", style="flow", data=df)
plt.title("2 Hop Average Throughput")
plt.show()


sns.set_style("darkgrid")
sns.lineplot(x="queue_size", dashes=False, y="throughput", hue="flow", style="flow", data=df)
plt.title("2 Hop Average Throughput")
plt.xlim(0, 100)
plt.show()
```

```python
def plot(hop=2,queue=100):
    f_bbr = open(str(hop) + "hop_" + str(queue) + "_bbr", "r").readlines()
    f_bbr = f_bbr[6:-6]

    bbr = []
    for i in range(50):
        bbr.append(0)
    for line in f_bbr:
        speeds = line.split("ytes")[1].split()
        if (speeds[1]=='Mbits/sec'):
            speed = float(speeds[0])
        else:
            speed = float(speeds[0])/1000.0
        bbr.append(speed)
    for i in range(50):
        bbr.append(0)
        
    f_cubic = open(str(hop) + "hop_" + str(queue) + "_cubic", "r").readlines()
    f_cubic = f_cubic[6:-6]
    cubic = []
    for line in f_cubic:
        speeds = line.split("ytes")[1].split()
        if (speeds[1]=='Mbits/sec'):
            speed = float(speeds[0])
        else:
            speed = float(speeds[0])/1000.0
        cubic.append(speed)
    
    
    ids =[]
    data = []
    for i in range(301):
        data.append([i,cubic[i],'cubic'])
        data.append([i,bbr[i],'bbr'])
    
    df = pd.DataFrame(data = data,
                      columns=['time','speed','flow'])
    
    sns.set_style("darkgrid")
    sns.lineplot(x="time", dashes=False, y="speed", hue="flow", style="flow", data=df)
    plt.title(str(hop)+"-Hop Speeds, Queue Size=" +str(queue))
    plt.show()
    
    
plot()
```

```python
for i in (list(range(4,100,4))+list(range(100,3000,50))):
    plot(2,i)
```

```python
for i in range(40,320,40):
    plot(4,i)
```

```python
for i in range(40,320,40):
    plot(8,i)
```
